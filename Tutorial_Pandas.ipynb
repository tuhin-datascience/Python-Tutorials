### **What is Pandas?**

**Pandas is a Python library developed by Wes McKinney in 2008. It is one of the most fundamental libray of Python that is used in Data Analysis.** 
* Pandas create a DataFrame or Series for for data manipulation with integrated indexing.
* Pandas can read and write data in different file format such as CSV, TXT, Excel, HTML, JASON, SQL etc. 
* Pandas can effectively handle the missing data.
* Pandas is capable of reshaping and pivoting of data sets.
* Pandas is also capable of Label-based slicing, fancy indexing, and subsetting of large data sets.
* Pandas can effectively manage column insertion and deletion.
* Pandas also able to manage merging and joining
* Pandas provides data filtration
* Pandas is capable of grouping and aggregating the data. 
* Pandas is also effectively handle Time Series Analysis

### **More DataFrame Operation**

```python
# Create another DataFrame
data = {'Company':['TESLA','TESLA','TOYOTA','TOYOTA','GM','GM'],
       'Person':['Cian','Donal','Flynn','John','Neil','Lucas'],
       'Sales':[2200,1900,3540,3100,2840,3850]}
df = pd.DataFrame(data)
df
```

```python
# Adding a new column ro the DataFrame
df['Cost'] = [1500, 1400, 3000, 2800, 2500, 3000]
df
```

```python
# Adding a new column through calculation.
df['Profit'] = df.Sales - df.Cost
df
```

### **apply()**
Pandas.apply() allow the users to pass a function and apply it on every single value of the Pandas series. It comes as a huge improvement for the pandas library as this function helps to segregate data according to the conditions required due to which it is efficiently used in data science and machine learning.

```python
# Finding the length of the string
df['Company'].apply(len)
```

```python
# Adding a new column using apply()
df['Projected Profit'] = df['Cost'].apply(lambda x: x*0.35)
```

```python
df
```

**Pandas Groupby: Summarising, Aggregating, and Grouping data in Python**

```python
# Finding the sum of Profit for each of the companies

df.groupby('Company')['Profit'].sum()
```

```python
# Average Sales by the Sales Persons
df.groupby('Person')['Sales'].mean()
```

```python
# Grouping by multiple coulmns

df.groupby(['Company', 'Person'])['Sales'].count()
```

```python
# Useage of multiple aggregate functions

df.groupby('Company', as_index=False).agg({'Profit':'sum', 'Sales':'sum', 'Cost':'mean'})
```

### **Joining DataFrame [Inner, Left & Right]** 

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
left
```

```python
right = pd.DataFrame({'key': ['K0', 'K1', 'K4', 'K5'],
                          'C': ['C0', 'C1', 'C4', 'C5'],
                          'D': ['D0', 'D1', 'D4', 'D5']})
right
```

```python
# Inner join based on a column

inner = pd.merge(left, right, how='inner', on='key')
inner
```

```python
left1 = pd.merge(left, right, how='left', on='key')
left1
```

```python
# Right join based on a column
right1 = pd.merge(left, right, how='right', on='key')
right1
```

### **Using iloc and loc to select rows and columns in Pandas DataFrames**

```python
df =pd.DataFrame({'Country':['Ireland', 'United Kingdom', 'France','Holland', 'Germany', 'Spain', 'Portugal', 'Italy'],
              'GDP':[75000, 65000, 67500, 67000, 68000, 55000, 50000, 40000],
              'Life Expectancy':[75.0, 72.5, 74.5, 76, 74, 73, 74, 70],
              'Pop': [5,80, 90, 25, 96, 35, 40, 45],
              'Year': [2019,2019,2019,2019,2019,2019,2019,2019]    
})
df
```

* **Selecting data by row numbers (.iloc)**

```python
# The first record
df.iloc[0]
```

```python
# The last record
df.iloc[-1]
```

```python
# All rows of the first column
df.iloc[:,0]
```

```python
#First three record
df.iloc[0:3]
```

```python
# All rows of first two columns
df.iloc[:, :2]
```

```python
# Row 1,3,5 and column 0,2,4
df.iloc[[1,2,3], [0, 2,4]]
```

* **Selecting data by label or by a conditional statement (.loc)**

```python
df.head()
```
