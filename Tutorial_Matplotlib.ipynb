# **Matplotlib Tutorial**

## **What is Matplotlib?**
Matplotlib is one of the oldest and the most popular plotting library in Python. First version od Matplotlib was relesed in 2003. It is derived from similar fashion with Matlab. It is highly compatible with Numpy and Pandas etc. We can perform a different kind of task such as 

*  Visulize the data
*  Identify Trends
*  Identify outliers
*  Highlighting a point, etc.


### **Anatomy of Matplotlib**

Matplotlib has a few key components that helps us to create a well defined graphs and we will discuss them one by one.

*  Figure - The figure is what contains all of the other elements of the plot. You can think of it
as the canvas that all of the elements of the plot are painted on.
*  Axes - Plots have X and Y axes, with one variable located on the X-axis and one variable on
the Y-axis.
*  Title - The title is the description given to the plot.
*  Legend - contains information regarding what the various symbols within the plot represent.
*  Ticks - Ticks are small lines used to point to different regions of the graph, mark specific items,
or delineate different thresholds. For example, if the X-axis of a graph contains the values 0 to
100, ticks may show up at 0, 20, 40, 60, 80, and 100. Ticks run along the sides, as well as the
bottom, of the graph.
*  Grids - Grids are lines in the plots background that make it easier to distinguish where
different values on the X and Y axes intersect.
*  Lines/Markers - Lines and markers are what represent the actual data within a plot. Lines
are typically used to graph continuous values, while markers/points are used to graph discrete
values.

```python
x = np.array([1,2,3,4,5])
y = np.square(x)

plt.plot(x,y)
plt.show()
```

As we can see, we have set the value of x and y that is square of x and then plot the same.
Next, we defined anoter variable that is cube of x and plot the same in the same graph. 

```python
x = np.array([1,2,3,4,5])
y = np.square(x)
z = x**3

plt.plot(x,y, z)
plt.show()
```

There a few problem with this graph.
*  Size is too small
*  No label is there
*  No title is there
*  No legend is defined, etc. 


Lets fix all these problems one by one. 

```python
x = np.array([1,2,3,4,5])
y = np.square(x)
z = x**3

# Fix the size of the graph
plt.figure(figsize=(16, 8))

# Set a title of the graph
plt.title('Title: X vs X-Square and X Cube')

# Set a caption for the X axis.
plt.xlabel('Value of X')

#Set a value for y-axis.
plt.ylabel('Value of Y and Z')

# Lets chnage the colour and style for each of these lines
plt.plot(x, color = 'royalblue',  label='X')
plt.plot(y, color = 'firebrick', linestyle = '-.', label='X Squared')
plt.plot(z, color = 'teal', linestyle = '--', label='X Cubed')

# Set a legend
plt.legend(loc='upper left', fontsize=18)
plt.show()
```

### **Subplot**

Axes objects sit within the figure you have created. Creating an axes object will give you greater
control over how data is visualized and other elements of the plot are created.
When using an axes object, you can control how individual subplots are displayed on those axes.
You can think of it like this: Figures hold axes and every axes object can store its own plots. The Axes
instance will contain most of the elements of a figure and you can have multiple Axes for a single
figure.

As we've seen previously, by default this creates a standard axes object that fills the entire figure. plt.axes also takes an optional argument that is a list of four numbers in the figure coordinate system. These numbers represent `[left, bottom, width, height]` in the figure coordinate system, which ranges from 0 at the bottom left of the figure to 1 at the top right of the figure.

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
names = ['2019', '2020', '2021']
values = [19, 50, 30]

ax.bar(names, values)
plt.show()
```

Matplotlib allows you to create multiple plots within the same figure. In order to add multiple plots,
you need to create a subplot for each plot in the figure youd like to use.
This is done with the add_subplot() function, which accepts a series of numeric arguments.
The first number specifies how many rows you want to add to the figure, the second number specifies
how many columns you want to add, and the third number specifies the number of the plot that
you want to add.
This means that if you in passed in 111 into the add_subplots()function, one new subplot would
be added to the figure. Meanwhile, if you used the numbers 221, the resulting plot would have four
axes with two columns and two rows - and the subplot youre forming is in the 1st position.

```python
fig = plt.figure()


names = ['2019', '2020', '2021']
values = [19, 50, 30]
value2 = [49, 20, 40]

ax = fig.add_subplot(121)  # 1 Row 2 Columns Figure 1 => 121
ax2 = fig.add_subplot(122) # 1 Row 2 Columns Figure 2 => 122

ax.bar(names, values)
ax2.bar(names, value2)

plt.show()
```

Lets say, we want to create a figure with 2 rows and 3 columns 

```python
fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(15, 4))
ax[0, 0].text(.4, .5, "axes[0, 0]")
ax[0, 1].text(.4, .5, "axes[0, 1]")
ax[0, 2].text(.4, .5, "axes[0, 2]")
ax[1, 0].text(.4, .5, "axes[1, 0]")
ax[1, 1].text(.4, .5, "axes[1, 1]")
ax[1, 2].text(.4, .5, "axes[1, 2]")
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

names = ['2019', '2020', '2021']
values = [19, 50, 30]
values2 = [49, 20, 40]

ax[0].bar(names, values)
ax[0].set_title('Google')


ax[1].bar(names, values2)
ax[1].set_title('Facebook')

# Set common labels
fig.text(0.5, 0.04, 'Years', ha='center', va='center')
fig.text(0.06, 0.5, 'Revenue in Billions', ha='center', va='center', rotation='vertical')

plt.suptitle('Annual Revenue', fontsize=14)
#plt.xlabel('')
#plt.ylabel('')

plt.show()
```

### **Tick**

We can edit the position and labels of your plots ticks. With the set_xticks() and set_yticks() functions, we can pass in lists of the positions where we want the ticks to be displayed.

Next, we can use set_xlabels() and set_ylabels() to label the ticks as we want. There are a variety of other options for customizing ticks as well.

```python
import math

a = np.arange(0, 2*math.pi, 0.25)
sine = np.sin(a)
fig = plt.figure(figsize=(7,7))
ax = fig.add_axes([0.1, .1, 0.75, 0.75])
ax.plot(a, sine)

ax.set_xlabel('X-Axis')
ax.set_ylabel('Y-Axis')
ax.set_title('Sine Curve')

ax.set_xticks([0,2,4,6,8])
ax.set_xticklabels(['zero', 'two', 'four', 'six', 'eight'])
ax.set_yticks([-1, 0, 1])

plt.show()
```

### **3D plots in Mathplotlib**

In Matplotlib we can generate 3D plots as well, here we can see some of the examples of 3D plots.

```python
from mpl_toolkits import mplot3d
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
```

```python
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python


fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='viridis')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');

```

```python
def f(x, y):
    return (np.sin(np.sqrt(x ** 2 + y ** 2)) + (np.sqrt(x ** 2 + y ** 2)))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='viridis')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

ax.view_init(60, 35) #60 degrees above the x-y plane) and an azimuth of 35 degrees;
```
