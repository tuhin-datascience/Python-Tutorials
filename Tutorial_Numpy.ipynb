```python
# Dot product and Matrix Maths
# Note: when we multiply in Numpy, we use elementwise multiplication, not matrix multiplication

a = np.array([1,2])
b = np.array([6,8])

c1 = a.dot(b)
print(c1)
print("\n")
c2 = np.dot(a,b)
print(c2)

# c = (1*6) + (2*8) = 6 + 16 = 22
```

```python
a = np.array([[1,2], [4,6]])
b = np.array([[4,1], [2,3]])

c= np.dot(a,b)

print(c)
```

```python
a = np.array([[1,2], [4,6]])
b = np.array([[4,1], [2,3]])

c= np.matmul(a,b)
print(c)
```

### **Random Number in Numpy**
Creating pseudo random number is very important aspect of Numpy and higly used in machine learning algorithm.

```python
# 1D random numbers
r1 = np.random.rand(10)
print(r1)
```

```python
# 2D random numbers
r2 = np.random.rand(5,2)
print(r2)
```

The pseudo random number generator is a mathematical function that generates a sequence of nearly random numbers. It takes a parameter to start off the sequence, called the seed. The function is deterministic, meaning given the same seed, it will produce the same sequence of numbers every time. The choice of seed does not matter.

```python
import random
random.seed(0) 
print(np.random.rand(4))
```

```python
random.seed(0) 
perm = np.random.permutation(10) 
print(perm) 
```

```python
# Random numbers that are normally distributed
r4 = np.random.randn(10)
print(r4)
```

```python
print(r4.mean())
print("\n")
print(r4.std())
```

As we can the mean is close to zero and the standard deviation is close to 1.

```python
# Random interger number generation between 1 to 6
r5 = np.random.randint(1,7)
print(r5)
```

```python
# Random interger number generation between 1 to 6 for two dices

r6 = np.random.randint(1,7, 2)
print(r6)
```

### **Iterator in Numpy**

### **Horizontal and Vertical Split**

```python
a = np.arange(30).reshape(2,15)
print(a)
```

```python
# splitting the arrary into 3 parts horizontally
h = np.hsplit(a,3)
```

```python
print(h[0])
print("\n")
print(h[1])
print("\n")
print(h[2])
print("\n")
```

```python
a = np.arange(30).reshape(2,15)
print(a)
```

```python
v = np.vsplit(a, 2)
```

```python
print(v[0])
print("\n")
print(v[1])
```

### **As we discussed before that Numpy array computes very fast infact it computes faster than normal Python list.**

* It is faster than Python list
* It takes less memomy than Python list
* It is more convenient than Python list


```python
import time
import sys
```

```python
# Define a variable x 
x = 100000
```

```python
# Arange() is a function that creats a series of Numpy array.
# Here we create two arrays
array1 = np.arange(x)
array2 = np.arange(x)
```

```python
print(array1)
print("\n")
print(array2)
```

```python
# Here, we add two Numpy array and check the requires to complete the operation.
starttime = time.time()
result = array1 + array2
endtime = (time.time() - starttime)*1000     #time in milisecond
print(endtime)
```

We will do the same operation with the Python lists now.

```python
# Here, we define two Python list
list1 = range(x)
list2 = range(x)
```

```python
# Here, we add two Python list and check the requires to complete the operation.
starttime = time.time()
result = [(x+y) for x,y in zip(list1, list2)]
endtime = (time.time() - starttime)*1000 #time in milisecond
print(endtime)
```

As we can see, the operation on Numpy array is much dfaster than the Python list.
